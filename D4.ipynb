```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 500
M = 500
t = 5000
T = 0.005
h = 1 / max(N, M)
dt = T / t
u = np.zeros((t, N, M))
u[0, 0] = np.zeros(M)
u[0, N-1] = np.zeros(M)
u[0, :, 0] = np.zeros(N)
u[0, :, M-1] = np.zeros(N)
```

```python
def f(x, y):
    A = (x - 0.5) ** 2 - (y - 0.5) ** 2 < 0.02**2 
    B = (x - 0.5) ** 2 - (y - 0.5) ** 2 > 0.015**2
    return 50 *  np.array(A*B)
```

```python
X, Y = np.meshgrid(np.linspace(0, 1, M, endpoint=False), np.linspace(0, 1, N, endpoint=False))
plt.imshow((f(X, Y)),)
```

```python
for i in range(1,t):
    u[i, 1:N-1, 1:M-1] = u[i-1, 1:N-1, 1:M-1] +dt * (f(X[1:N-1, 1:M-1], Y[1:N-1, 1:M-1]) + (u[i-1, 2:N, 1:M-1] + u[i-1, 0:N-2, 1:M-1]+u[i-1, 1:N-1, 2:M] + u[i-1, 1:N-1, 0:M-2] - 4 * u[i-1, 1:N-1, 1:M-1]) / h**2)
```
